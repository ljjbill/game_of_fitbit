## Import Dependencies

```python
import pandas as pd
```

```python
def toDf(data):
    df = pd.read_csv(data)
    return df
```

## DailyActivities Merge

```python
data = '../Excels/Fitabase Data 3.12.16-4.11.16/dailyActivity_merged.csv'
data2 = '../Excels/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv'

df = toDf(data)
df2 = toDf(data2)
df.head()
```

```python
df2.head()
```

```python
dailyActivity = pd.concat([df, df2], ignore_index=True)
dailyActivity.head()
```

```python
dailyActivity.to_csv('dailyActivity.csv', sep=',', encoding='utf-8')
```
