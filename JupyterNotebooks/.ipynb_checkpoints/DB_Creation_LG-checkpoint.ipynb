```python
Base = declarative_base()
```

```python
class dailyActivity(Base):
    __tablename__= 'dailyActivity'
    index = Column(VARCHAR, primary_key=True)
    Id = Column(VARCHAR)
    ActivityDate = Column(VARCHAR)
    TotalSteps = Column(Integer)
    TotalDistance = Column(Integer)
    TrackerDistance = Column(Integer)
    LoggedActivitiesDistance = Column(Integer)
    VeryActiveDistance = Column(Integer)
    ModeratelyActiveDistance = Column(Integer)
    LightActiveDistance = Column(Integer)
    SedentaryActiveDistance = Column(Integer)
    VeryActiveMinutes = Column(Integer)
    
```
